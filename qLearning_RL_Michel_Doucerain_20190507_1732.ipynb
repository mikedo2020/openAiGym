```
import gym
import numpy as np
```

```
env = gym.make('FrozenLake-v0')

learning_rate = .7
discount_rate = .9
episodes = 500


def qLearn(env, episodes, discount_rate, learning_rate):
  rewards = []
  Q = np.zeros([env.observation_space.n,env.action_space.n])
  for i in range(episodes):
      state = env.reset()
      total_reward = 0
      done = False
      for t in range(10):
          env.render()
          action = np.argmax(Q[state,:] + np.random.randn(1,env.action_space.n)*(1./(i+1)))
          print("best action",action)
          next_state,reward,done,info = env.step(action)
          env.render()
          print(next_state,reward,done,info)
          Q[state,action] = Q[state,action] + learning_rate*(reward + discount_rate*np.max(Q[next_state,:]) - Q[state,action])
          total_reward += reward
          state = next_state
          print('state',state)
          if done:
              print('total_reward',total_reward)
              break
      if total_reward > 0:
        print("Won after {} episodes".format(i+1))
        print("Won after {} timesteps".format(t+1))
        break
      rewards.append(total_reward)
  print("Total Rewards " + str(sum(rewards)))
  print("Q values",Q)
  
```

```
qLearn(env, episodes, discount_rate, learning_rate)
```
